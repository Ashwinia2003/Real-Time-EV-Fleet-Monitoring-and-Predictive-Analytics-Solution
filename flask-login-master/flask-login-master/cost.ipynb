```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from a CSV file
df = pd.read_csv(r'C:\Users\ashwini\OneDrive\Desktop\Ev management\flask-login-master\flask-login-master\COMED_hourly.csv', parse_dates=['Datetime'], index_col='Datetime')

# Convert 'COMED_MW' column to numeric values (if necessary)
df['COMED_MW'] = pd.to_numeric(df['COMED_MW'], errors='coerce')

# Fill missing values (optional: you could also use df.dropna())
df['COMED_MW'] = df['COMED_MW'].fillna(0)

# Define the cost per MWh (e.g., $100 per MWh)
cost_per_MWh = 100

# 1. Calculate total energy consumption for a specific period (e.g., daily, weekly, or monthly)
# Resample to daily energy consumption (you can change to 'W' for weekly, 'M' for monthly, etc.)
df_resampled = df.resample('D').sum()  # Use 'D' for daily, 'W' for weekly, 'M' for monthly, etc.

# 2. Calculate the operational costs based on energy consumption
df_resampled['Operational Costs'] = df_resampled['COMED_MW'] * cost_per_MWh

# Display the result
print(df_resampled[['COMED_MW', 'Operational Costs']])

# Plotting the energy consumption and operational costs
plt.figure(figsize=(12, 6))

# Plot Energy Consumption
plt.subplot(1, 2, 1)
plt.plot(df_resampled.index, df_resampled['COMED_MW'], label='Energy Consumption (MW)', color='blue')
plt.title('Energy Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Energy Consumption (MW)')
plt.xticks(rotation=45)
plt.grid(True)

# Plot Operational Costs
plt.subplot(1, 2, 2)
plt.plot(df_resampled.index, df_resampled['Operational Costs'], label='Operational Costs', color='green')
plt.title('Operational Costs Over Time')
plt.xlabel('Date')
plt.ylabel('Operational Costs ($)')
plt.xticks(rotation=45)
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the plots as an image (for screenshots)
plt.savefig('energy_consumption_and_cost_analysis.png')

# If you want to save the result to a new CSV file
df_resampled.to_csv('energy_consumption_and_costs.csv')
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

# Load the dataset from a CSV file
df = pd.read_csv(r'C:\Users\ashwini\OneDrive\Desktop\Ev management\flask-login-master\flask-login-master\COMED_hourly.csv', parse_dates=['Datetime'], index_col='Datetime')

# Convert 'COMED_MW' column to numeric values (if necessary)
df['COMED_MW'] = pd.to_numeric(df['COMED_MW'], errors='coerce')

# Fill missing values (optional: you could also use df.dropna())
df['COMED_MW'] = df['COMED_MW'].fillna(0)

# Define the cost per MWh (e.g., $100 per MWh)
cost_per_MWh = 100

# 1. Calculate total energy consumption for a specific period (e.g., daily, weekly, or monthly)
# Resample to daily energy consumption (you can change to 'W' for weekly, 'M' for monthly, etc.)
df_resampled = df.resample('D').sum()  # Use 'D' for daily, 'W' for weekly, 'M' for monthly, etc.

# 2. Calculate the operational costs based on energy consumption
df_resampled['Operational Costs'] = df_resampled['COMED_MW'] * cost_per_MWh

# Display the result
print(df_resampled[['COMED_MW', 'Operational Costs']])

# Plotting the energy consumption and operational costs using Plotly

# Create figure
fig = go.Figure()

# Add Energy Consumption line
fig.add_trace(go.Scatter(x=df_resampled.index, 
                         y=df_resampled['COMED_MW'],
                         mode='lines', 
                         name='Energy Consumption (MW)',
                         line=dict(color='blue')))

# Add Operational Costs line
fig.add_trace(go.Scatter(x=df_resampled.index, 
                         y=df_resampled['Operational Costs'],
                         mode='lines', 
                         name='Operational Costs ($)',
                         line=dict(color='green')))

# Update layout
fig.update_layout(
    title="Energy Consumption and Operational Costs Over Time",
    xaxis_title="Date",
    yaxis_title="Value",
    legend_title="Metrics",
    template="plotly_dark",  # Optional: change the template
    xaxis=dict(tickformat="%Y-%m-%d"),  # Format x-axis ticks to show date
    height=600,
    width=1000
)

# Show the figure
fig.show()

# Optionally, save the figure as an image (requires orca or kaleido)
fig.write_image("energy_consumption_and_cost_analysis_plotly.png")

# If you want to save the result to a new CSV file
df_resampled.to_csv('energy_consumption_and_costs.csv')
```
